```python
import pandas as pd
import datetime as dt

df_tweets = pd.read_csv(r'tweets_with_sentiment_scores/bitcoin_tweets_with_sentiment_score_21_05_20_08.csv', delimiter = ",", error_bad_lines=False)
df_prices = pd.read_csv(r'prices_files/BTC-USD-01-04-20-08.csv', delimiter = ",")
# df_tweets = df_tweets.iloc[: , 2:]
stock = "Bitcoin"
df_tweets = df_tweets.dropna()

# df_tweets = df_tweets.sort_values(by='DATE')
# df_tweets['DATE'] = df_tweets['DATE'].dt.strftime('%d/%m/%Y')
# df_tweets['DATE'].unique()

df_tweets["sentiment_score"] = pd.to_numeric(df_tweets["sentiment_score"], downcast="float")
df_tweets = df_tweets.groupby('DATE') \
       .agg({'DATE':'count', 'sentiment_score':'mean'}) \
       .rename(columns={'DATE':'tweet count','sentiment_score':'sentiment_score'}) \
       .reset_index()

# rslt_df = df_tweets[df_tweets['DATE'] == '07/10/2021']
# rslt_df
```

```python
from datetime import datetime

# df_tweets["sentiment_score"] = pd.to_numeric(df_tweets["sentiment_score"], downcast="float")

# df_tweets["sentiment_score"].mean()
# df_tweets['average by day'] = df_tweets.groupby('DATE')['sentiment_score'].transform('mean')
# df_tweets['tweet count'] = df_tweets.groupby('DATE')['DATE'].transform('count')

# df_scores = df_tweets.drop_duplicates(subset=['DATE'])

df_scores = df_tweets[['DATE','sentiment_score', 'tweet count']]

df_count = df_tweets.drop_duplicates(subset=['DATE'])
df_count = df_count[['DATE','tweet count']]

df_prices = df_prices[['Date','Close']]
dates_and_scores = df_scores.values.tolist()
dates_and_prices = df_prices.values.tolist()
dates_and_counts = df_count.values.tolist()

print(dates_and_scores)
print(dates_and_counts)
# fixing month in place of day 08 in few cases
for date_and_score in dates_and_scores:
    if int(date_and_score[0].split("/")[0]) in range(0,13):
        month = date_and_score[0].split("/")[0]
        date = date_and_score[0].split("/")[1]
        proper_date = date+"/"+month+"/"+date_and_score[0].split("/")[2]
        date_and_score[0] = proper_date
    date_time_obj = datetime.strptime(date_and_score[0], '%d/%m/%Y')
    date_and_score[0] = date_time_obj

from operator import itemgetter
dates_and_scores = sorted(dates_and_scores, key=itemgetter(0))
for date_and_score in dates_and_scores:
    date_and_score[0] = str(date_and_score[0].strftime('%d/%m/%Y'))

for date in dates_and_prices:
    date[0] = date[0].replace("-","/")
    year = date[0].split("/")[0]
    day = date[0].split("/")[2]
    proper_date = day+"/"+date[0].split("/")[1]+"/"+year
    date[0] = proper_date

df_tweets.tail(20)
# print(dates_and_scores.index(dates_and_scores[10]))
# print(dates_and_prices.index(dates_and_scores.index(dates_and_scores[10])))

print(len(dates_and_prices), dates_and_prices ,'\n')

if stock=='Tesla':
    dates_and_scores.pop(0)
elif stock=='Gamestop':
    dates_and_scores.pop(0)
    dates_and_scores.pop(0)
    dates_and_scores.pop(0)

for date_and_score in dates_and_scores:
    if date_and_score[0] not in dates_and_prices[dates_and_scores.index(date_and_score)]:
        dates_and_prices.insert(dates_and_scores.index(date_and_score),
                                [date_and_score[0], dates_and_prices[dates_and_scores.index(date_and_score)-1][1]])

print(len(dates_and_scores), dates_and_scores)
print(len(dates_and_prices), dates_and_prices)
```

```python
#some plotting options are commented

import matplotlib.pyplot as plt
import numpy as np

f = plt.figure(figsize=(100,20))

dates = [date[0] for date in dates_and_scores]
scores = [score[1] for score in dates_and_scores]
prices = [price[1] for price in dates_and_prices]
counts = [score[2] for score in dates_and_scores]
#dates.reverse()
#scores.reverse()

rolling_score = 0
seven_days_average = 0
rolling_scores = []
seven_days_average_scores = []

for score in scores:
    rolling_score += score
    rolling_scores.append(rolling_score)

def get_moving_average(window_size):
    x_days_average_scores = []
    i = 0
    k = 0
    while i < len(scores) - window_size + 1:
        this_window = scores[i : i + window_size]
        window_average = sum(this_window) / window_size
        x_days_average_scores.append(window_average)
        i += 1
    while k<window_size-1:
        x_days_average_scores.insert(0, None)
        k+=1

    return x_days_average_scores

seven_days_average_scores = get_moving_average(7)
five_days_average_scores = get_moving_average(5)
ten_days_average_scores = get_moving_average(10)

print(len(scores))
print(len(seven_days_average_scores))
print(rolling_scores)

fig,ax = plt.subplots()

ax.plot(dates, prices, color="blue", marker="o", label='cena waloru')
ax.set_xlabel("data",fontsize=14)
ax.set_ylabel("cena",color="black",fontsize=14)
lines, labels = ax.get_legend_handles_labels()

ax2=ax.twinx()
ax2.plot(dates, scores,color="red",marker="o", label="ocena sentymentu")
ax2.set_ylabel("ocena sentymentu",color="black",fontsize=14)
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.grid(True)
title = stock+' - ocena sentymentu oraz cena'

# ax3=ax.twinx()
# ax3.plot(dates, rolling_scores,color="green",marker="o")
# ax3.set_ylabel("'rolujÄ…ca' ocena sentymentu",color="green",fontsize=14)
# title = stock+" - ocena sentymentu 'rolujÄ…ca' oraz cena"
# plt.title(title)

# ax7=ax.twinx()
# ax7.plot(dates, seven_days_average_scores,color="orange",label="siedmiodniowa Å›rednia kroczÄ…ca oceny sentymentu")
# ax7.set_ylabel("Å›rednia ocena sentymentu",color="orange",fontsize=14)
# ax7.grid(True)
# lines, labels = ax7.get_legend_handles_labels()
# plt.title(title)

# ax5=ax.twinx()
# ax5.plot(dates, five_days_average_scores,color="green", label="piÄ™ciodniowa Å›rednia kroczÄ…ca oceny sentymentu")
# ax5.set_ylabel("Å›rednia ocena sentymentu",color="black",fontsize=14)
# # ax5.get_yaxis().set_visible(False)
# title = stock+' - Å›rednie kroczÄ…ce oceny sentymentu oraz cena'
# ax5.grid(True)
# lines2, labels2 = ax5.get_legend_handles_labels()

# ax10=ax.twinx()
# ax10.plot(dates, ten_days_average_scores,color="indigo", label="dziesiÄ™ciodniowa Å›rednia kroczÄ…ca oceny sentymentu")
# ax10.get_yaxis().set_visible(False)
# lines3, labels3 = ax10.get_legend_handles_labels()


ax3 = ax.twinx()
ax3.bar(dates, height=counts, color="royalblue", alpha=0.3)
ax3.get_yaxis().set_visible(False)
ax3.set_ylim(top = max(counts)/0.15)

#for plotting legend
# ax.legend(lines + lines2 + lines3, labels + labels2 +labels3, bbox_to_anchor=(0, 1.12), loc='upper left', ncol=1, prop={'size': 8})
# ax.legend(lines + lines2, labels + labels2, bbox_to_anchor=(1, 0.98), loc='upper right', ncol=1, prop={'size': 9})
# ax.legend(lines + lines2 + lines3,labels + labels2 +labels3, loc='best')
ax.legend(lines + lines2,labels + labels2, loc='best')

ax.set_xticks(dates[::5])
ax.set_xticklabels(dates[::5], minor=False, rotation=90)

figure = plt.gcf() # get current figure
figure.set_size_inches(17, 7)
plt.title(title)

# plt.savefig('zdjÄ™cia i skrinshoty/'+title+'.jpg', bbox_inches='tight', dpi=100)
plt.show()

```
