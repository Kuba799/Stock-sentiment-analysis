```python
import pandas as pd
import re
from preprocessor import clean, set_options
df_tweets = pd.read_csv(r'tweets_from_snscrape/tesla_snscrape_2021-08-01_2021-10-29.csv', delimiter = ",", encoding="utf-8")
df_tweets.columns = ["DATE", "TWEET"]
pd.options.display.max_colwidth = 3000
```

```python
df_tesla = df_tweets[0:1]
df_tesla = df_tesla.append(df_tweets[6:8])
df_tesla = df_tesla.append(df_tweets[11:12])
df_tesla = df_tesla.append(df_tweets[47:48])
df_tesla = df_tesla.append(df_tweets[45:46])
# df_tesla = df_tesla.append(df_tweets[564:565])
df_tweets = df_tesla
df_tweets
```

```python
tweets_list = []
for index, row in df_tweets.iterrows():
     tweets_list.append([row.DATE, row.TWEET])
# removing b'' or b""
df_tweets = df_tweets.replace(to_replace ="^b'", value = "", regex = True)
df_tweets = df_tweets.replace(to_replace ='^b"', value = "", regex = True)
df_tweets = df_tweets.replace(to_replace ="'\s*$", value = "", regex = True)
df_tweets = df_tweets.replace(to_replace ='"\s*$', value = "", regex = True)
df_tweets = df_tweets.replace(to_replace ='\\n', value = "", regex = True)
df_tweets = df_tweets.replace(to_replace ='\#', value = "", regex = True)
df_tweets
```

```python
import preprocessor as p

tweets_list = []
for index, row in df_tweets.iterrows():
    tweets_list.append([row['DATE'], row['TWEET']])
# # # Print the updated dataframe
tweets_list_copy = tweets_list.copy()
p.set_options(p.OPT.URL, p.OPT.MENTION)
for tweet, tw in zip(tweets_list_copy, tweets_list):
    tw[-1] = p.clean(tweet[-1])
    tw[-1] = tweet[-1].replace("\\n", "")
```

```python

df_tweets = pd.DataFrame(tweets_list)
df_tweets.columns = ["DATE", "TWEET"]
df_tweets['DATE'] = pd.to_datetime(df_tweets['DATE'])
df_tweets['Date'] = df_tweets['DATE'].dt.strftime('%d/%m/%Y')
df_tweets['Time'] = df_tweets['DATE'].dt.strftime('%H:%M:%S')
df_tweets = df_tweets.drop("DATE",1)
df_tweets = df_tweets.rename(columns={'Date': 'DATE', 'Time': 'TIME'})
df_tweets
```
